```python
import numpy as np
import sys

sys.path.append(r"C:\Users\leila\Documents\Visual Studio\pppl_xics_2026\mirxics_jax")
sys.path.append(r"C:\Users\leila\Documents\Visual Studio\pppl_xics_2026\xics_ml_pipeline")

import xics_jax
from xicsrt_multiline_voigt import voigt, multiline_voigt, multiline_voigt_cdf_tab, multiline_voigt_random
```

```python
# loading Ar16+ spectrum setup
spectrum_config = xics_jax.load_spectrum_config('ar16', use_te=False)

params = xics_jax.ModelParams.from_dict(
    {
        'ti': 2.0,
        'scale_factor': 1.0,
        'w_intensity': 1.0,
        'x_intensity': 1.0,
        'y_intensity': 1.0,
        'z_intensity': 1.0,
    },
    spectrum_config,
)

# Wavelength grid used by xics_jax
wavelengths = np.linspace(3.94, 4.00, 400)

# Get per-line Voigt parameters
lines = xics_jax.evaluate_lines(wavelengths, params, spectrum_config)

# Convert xics_jax output into multiline_voigt_random inputs
line_locations = lines.location.values
line_intensities = lines.intensity.values
sigmas = lines.sigma.values
gammas = lines.gamma.values

# Sample wavelengths from the full Ar16 multiline Voigt distribution
sampled_wavelengths = multiline_voigt_random(
    line_locations,
    line_intensities,
    sigmas,
    gammas,
    size=10000,
    gridsize=None,
    cutoff=1e-4,
)
```
