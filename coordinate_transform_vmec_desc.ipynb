```python
import numpy as np
from desc.vmec import VMECIO
from desc.grid import Grid

# woutfile = '/u/npablant/data/w7x/vmec/w7x_ref_172/wout.nc'
woutfile = r"C:\Users\leila\Documents\Visual Studio\pppl_xics_2026\wout.nc"

# 1. Load the VMEC wout file into a DESC Equilibrium object
eq = VMECIO.load(woutfile)
```

```python
flux_coord_in = np.array([[0.2, 0.0, 0.2]])
xyz_coord = eq.map_coordinates(
    flux_coord_in, 
    inbasis=("rho", "theta", "zeta"),
    outbasis=("X", "Y", "Z"),
    )
flux_coord_out = eq.map_coordinates(
    xyz_coord, 
    inbasis=("X", "Y", "Z"),
    outbasis=("rho", "theta", "zeta"),
    )

eq.solve()

xyz_coord_solve = eq.map_coordinates(
    flux_coord_in, 
    inbasis=("rho", "theta", "zeta"),
    outbasis=("X", "Y", "Z"),
    )
flux_coord_solve = eq.map_coordinates(
    xyz_coord, 
    inbasis=("X", "Y", "Z"),
    outbasis=("rho", "theta", "zeta"),
    )
```

```python
# Set the global print format for floats to 12.6f
with np.printoptions(formatter={'float_kind': '{:12.6f}'.format}):
    print(flux_coord_in)
    print(flux_coord_out)
    print(flux_coord_solve)
    print('')
    print(xyz_coord)
    print(xyz_coord_solve)
```
